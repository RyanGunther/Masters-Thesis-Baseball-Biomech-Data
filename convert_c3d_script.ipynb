```python
# convert all C3D files to CSV (flat layout)
import os, glob, ezc3d, pandas as pd
from tqdm import tqdm

INPUT_DIR  = "/content/openbiomechanics/baseball_hitting/data/c3d"
OUTPUT_DIR = "/content/ALL_OBP_FILES"
os.makedirs(OUTPUT_DIR, exist_ok=True)

for c3d_path in tqdm(
        glob.glob(os.path.join(INPUT_DIR, "**", "*.c3d"), recursive=True),
        desc="Converting", unit="file"):
    c3d = ezc3d.c3d(c3d_path)
    pts, labels = c3d["data"]["points"], c3d["parameters"]["POINT"]["LABELS"]["value"]

    data = {f"{m}_{ax}": pts[ax_i, i, :]
            for i, m in enumerate(labels)
            for ax_i, ax in enumerate("XYZ")}

    out_name = os.path.splitext(os.path.basename(c3d_path))[0] + ".csv"
    pd.DataFrame(data).to_csv(os.path.join(OUTPUT_DIR, out_name), index=False)
```
